```python
# def get_log_prob(logits,token_id):
#   probabilities = torch.nn.functional.softmax(logits, dim=-1)
#   log_probabilities = torch.log(probabilities)
#   token_log_prob = log_probabilities[0][token_id]
#   return token_log_prob.item()
def get_log_prob(logits, token_id):
    probabilities = torch.nn.functional.softmax(logits, dim=-1)  # Convert logits to probabilities
    log_probabilities = torch.log(probabilities)  # Compute the log probabilities
    token_log_prob = log_probabilities[token_id]  # Index directly with token_id
    return token_log_prob.item()  # Return the scalar value


```

```python
import numpy as np
```

```python
def greedy_search(input_ids,length=5):
    if length == 0 :
      return input_ids

    output=model(input_ids)
    predictions=output.logits
    logits=predictions[0,-1,:]
    token_id=torch.argmax(logits).unsqueeze(0)

    token_score=  get_log_prob(logits,token_id)

    new_input_ids=torch.cat([input_ids,token_id.unsqueeze(0)],dim=-1)

    next_token=tokenizer.decode(token_id,skip_special_tokens=True)

    new_token={}

    new_token['token'] = next_token + f"-> {length}"
    new_token['token_score']=  np.exp(token_score)*100

    generated_tokens.append(new_token)
    input_ids=greedy_search(new_input_ids,length-1)
    return input_ids
```

```python
import os
import random
from langchain_community.document_loaders import DirectoryLoader, TextLoader
```

```python

# Define the path to the dataset in Colab
base_directory = "/content"  # This is where your SANAD dataset folders are located

# List of subdirectories corresponding to the categories
categories = ['Culture', 'Finance', 'Medical', 'Politics', 'Religion', 'Sports', 'Tech']

# Load documents from all categories
documents = []
for category in categories:
    category_path = os.path.join(base_directory, category)

    # Get all file paths in the category
    all_files = [os.path.join(category_path, f) for f in os.listdir(category_path) if f.endswith('.txt')]

    # Randomly sample 25% of the files
    sample_size = int(len(all_files))
    sampled_files = random.sample(all_files, sample_size)

    # Load only the sampled files
    for file_path in sampled_files:
        loader = TextLoader(file_path)
        documents.extend(loader.load())

# Display the content of the first few documents
for doc in documents[:5]:
    print(f"File: {doc.metadata['source']}")
    print(doc.page_content[:500])  # Print first 500 characters for brevity
    print("\n" + "="*50 + "\n")
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
import bs4
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=3000,chunk_overlap=200,add_start_index=True)
```

```python
all_splits=text_splitter.split_documents(documents)
```

```python
# Checking the number of splits and some sample splits
print(f"Total splits: {len(all_splits)}")

# Display the content of the first few splits
for i, split in enumerate(all_splits[:5]):
    print(f"Split {i + 1}:")
    print(split.page_content[:500])  # Print the first 500 characters for brevity
    print(f"Start Index: {split.metadata['start_index']}")
    print("\n" + "="*50 + "\n")
```

```python
from sentence_transformers import SentenceTransformer, util
from langchain_chroma import Chroma
```

```python
class embedding:
  def __init__(self):
    self.model = SentenceTransformer('paraphrase-multilingual-MiniLM-L12-v2')

  def embed_documents(self, docs):
      embeddings= self.model.encode(docs)
      return embeddings.tolist()

  def embed_query(self, query):
    return self.model.encode(query).tolist()
```

```python
embed_model = embedding()
```

```python
result =embed_model.embed_query("hello")
len(result)
```
